### Combine dataset

```python
air_quality_no2 = pd.read_csv("air_quality_no2_long.csv",
                                  parse_dates=True)
```

```python
air_quality_no2 = air_quality_no2[["date.utc", "location",
                                       "parameter", "value"]]
```

```python
print(air_quality_no2.shape)
air_quality_no2.head()
```

```python
air_quality_pm25 = pd.read_csv("air_quality_pm25_long.csv",
                                   parse_dates=True)
```

```python
air_quality_pm25 = air_quality_pm25[["date.utc", "location",
                                         "parameter", "value"]]
print(air_quality_pm25.shape)
air_quality_pm25.head()
```

```python
air_quality = pd.concat([air_quality_pm25, air_quality_no2], axis=0)
print(air_quality.shape)
air_quality.head()
```

```python
air_quality = air_quality.sort_values("date.utc")
air_quality.head()
```

```python
# Concate the particular key data

air_quality_ = pd.concat([air_quality_pm25, air_quality_no2],
                          keys=["PM25", "NO2"])
print(air_quality_.shape)
air_quality_.head()
```

### Joins (Like in DBMS)

```python
stations_coord = pd.read_csv("air_quality_stations.csv")
air_quality = pd.merge(air_quality, stations_coord,
                           how='left', on='location')
air_quality.head()
```
